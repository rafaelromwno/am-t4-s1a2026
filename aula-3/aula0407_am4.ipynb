# Análise exploratórias de anomalias (Alunos)

### Possíveis idades negativas - (Incogruências ou Impossibilidades Lógicas) | Resultante: Normalização 

```python
base_credit.loc[base_credit['age'] < 0]
```

### Possíveis ganhos negativos - (Incongruências ou Impossibilidades) | Resultante: Nulos

```python
base_credit.loc[base_credit['income'] < 0]
```

### Possíveis valores nulos - (Dados nulos em base) | Resultante: Normalização

```python
base_credit.isnull().sum()
```

```python
base_credit[base_credit.isnull().any(axis=1)]
```

### Possíveis Empréstimos negativos - (Incongruências ou Impossibilidades) | Resultante: Nulo

```python
base_credit.loc[base_credit['loan'] < 0]
```
